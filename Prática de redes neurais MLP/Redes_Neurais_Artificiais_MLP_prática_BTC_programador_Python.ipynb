```python
import pandas as pd
```

```python
data = pd.read_csv("/content/credit_data.csv")
data.head()
```

```python
data.dropna(inplace=True)
```

```python
X_credit = data[['income', 'age', 'loan']]
Y_credit = data['default']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
X_credit = standard_scaler.fit_transform(X_credit)
```

```python
X_credit
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_credit, Y_credit, test_size = 0.25, random_state =0)
```

```python
X_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
credit_neural_network = MLPClassifier(verbose = True, max_iter=1000, tol = 0.000010, hidden_layer_sizes= (20,100))
credit_neural_network.fit(X_train, Y_train)
```

```python
predictions = credit_neural_network.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(Y_test, predictions)
```

```python
report = classification_report(Y_test, predictions, digits=4, zero_division='warn')
print(report)
```
